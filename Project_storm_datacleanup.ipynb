```python
import pandas as pd
from sqlalchemy import create_engine 
import datetime as dt
```

```python
# Importing csv

storm_csv = pd.read_csv('data/Storms_2020.csv')
storm_csv.head()
```

```python
storm_csv.columns
```

```python
# selecting the columns we want

new_stormcsv = storm_csv[['event_id', 'state', 'event_type','event_begin_time',
       'event_timezone', 'event_end_time','injuries_direct',
       'injuries_indirect', 'deaths_direct', 'deaths_indirect',
       'damage_property', 'damage_crops','magnitude','event_latitude',
       'event_longitude']]
new_stormcsv.head()
```

```python
#identifying types for each column
new_stormcsv.info()
```

```python
# cleaning date columns

new_stormcsv['event_begin_time'] = pd.to_datetime(new_stormcsv['event_begin_time'])
new_stormcsv['event_end_time'] = pd.to_datetime(new_stormcsv['event_end_time'])
```

```python
new_stormcsv.info()
```

```python
# Combining injury and death columns to be one

new_stormcsv['Injuries'] = new_stormcsv['injuries_direct'] + new_stormcsv['injuries_indirect']
new_stormcsv['Deaths'] = new_stormcsv['deaths_direct'] + new_stormcsv['deaths_indirect']
# new_stormcsv['event_month'] = dt.datetime.strftime(new_stormcsv['event_begin_time'], "%m")
new_stormcsv['event_month'] = new_stormcsv['event_begin_time'].dt.month
new_stormcsv.tail(30)
```

```python
new_stormcsv.columns
```

```python
clean_storm = new_stormcsv[['event_id', 'state', 'event_type', 'event_month', 'event_begin_time', 'event_timezone',
       'event_end_time', 'Injuries', 'Deaths', 'damage_property', 'damage_crops',
       'magnitude', 'event_latitude', 'event_longitude']]
clean_storm.head()
```

```python
# column renamings

clean_storm = clean_storm.rename(columns={'Injuries':'total_injuries', 'Deaths':'total_deaths'})
clean_storm.head()
```
